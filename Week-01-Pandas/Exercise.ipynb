```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn. 

brooklyn_condition = df['neighbourhood_group'] == 'Brooklyn'# create brooklyn condition filter

df_brookyln = df[brooklyn_condition] #the dataframe for brooklyn

id_nGroup_price_cols = ['id','neighbourhood_group','price'] # cols with just id and neighbourhood_group

df_brookyln[id_nGroup_price_cols] # display df_brookyln with just id and neighbourhood_group

```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.

name_of_host = ['Michael', 'David', 'John', 'Daniel'] #list with selected host name 

host_condition = df.host_name.isin(name_of_host) # check is list name_of_host is in col host

df_host = df[host_condition] # creating a data frame that filters name_of_host

host_cols = ['id','host_name']

df_host[host_cols] # display df with just id, host_name (host_cols)

```

```python
# 5. Create a new column called `adjusted_price` 
# that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before.

#***I have two solution***
#***solution 1 using filtering***
# filter williamsburg and then add 100 to that price

williamsburg_condition = df['neighbourhood'] == 'Williamsburg'# filter for just williamsburg

df_williamsburg= df[williamsburg_condition] # create a df for williamsburg_condition

df_williamsburg['adjusted_price'] = df_williamsburg['price'] + 100 # add column to df that will take col of price from df_williamsburg and add 100

df_williamsburg.head() #diplay the first five rows of the columns






```

```python
# 5. Create a new column called `adjusted_price` 
# that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before.

#***Solution 2 using np.where***

wiliamsburg_price_change = df['neighbourhood'] == 'Williamsburg'# create a filter for showing just Williamsburg

df['adjusting_price'] = np.where(wiliamsburg_price_change, df['price']+100,"no change") # adding col, checking for williamsburg condition, if true add 100 to price, if false "no change"

df[wiliamsburg_price_change] #display df with just williamsburg_price_change
```

```python
# 6. What % of the rooms are private

private_condition = df['room_type'] == 'Private room' # filter for private room

df_private = df[private_condition] # create a data frame with 

# total rows(listings) = 46527
percent_private = (df_private['room_type'].value_counts() / 46527) * 100 # get the value of the df_private col room type. divide that value with total and mulitply with 100

percent_private #show the percent
```

```python
# 6. what % of the rooms are shared.  

shared_condition = df['room_type'] == 'Shared room'# filter out the 'Shared room'

df_shared = df[shared_condition] #create a data frame with just the shared_condition

# total for shared = 987
precent_shared = (df_shared['room_type'].value_counts() / 46527)*100 #get the percent with (.value_counts() / total_listings)*100 

precent_shared # show the percent

```
